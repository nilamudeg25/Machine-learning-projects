```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
```

```python
#col=['pregnant','glocose','bp','skin','insulin','bmi','predigree','age']
df1=df.iloc[:,:-1]
df1[:]=scaler.fit_transform(df1[:])
df1.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=df1[['pregnant','glocose','bp','skin','insulin','bmi','predigree','age']]
y=df['target']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.40,random_state=20)
```

```python
print('shape of x_train:',x_train.shape)
print('shape of y_train:',y_train.shape)
print('shape of x_test:',x_test.shape)
print('shape of y_test:',y_test.shape)
```

# Gradient Boosting Classification model

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
gb_cls=GradientBoostingClassifier()
```

```python
gb_cls.fit(x_train,y_train)
```

```python
gb_pred = gb_cls.predict(x_test)
gb_pred
```

```python
from sklearn.metrics import accuracy_score
```

```python
print(accuracy_score(y_test,gb_pred))
```

# Ada Boosting Classification model

```python
from sklearn.ensemble import AdaBoostClassifier
ada_clf = AdaBoostClassifier()
```

```python
ada_clf.fit(x_train,y_train)
ada_pred = ada_clf.predict(x_test)
ada_pred
```

```python
print(accuracy_score(y_test, ada_pred))
```

# Extreme Gradient Boosting model

```python
from xgboost import XGBClassifier
```

```python
xgb_cls=XGBClassifier(booster='gbtree')
xgb_cls.fit(x_train,y_train)
```

```python
xgb_pred=xgb_cls.predict(x_test)
xgb_pred
```

```python
print(accuracy_score(y_test, xgb_pred))
```
