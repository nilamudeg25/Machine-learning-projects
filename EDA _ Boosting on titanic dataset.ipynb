# Use Encoding and encode all the categorical data

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
df1=df.drop(['Name','Ticket'],axis=1)
df1.head()
```

```python
data_encoding=df1.select_dtypes('object').columns
data_encoding
    
```

```python
for i in data_encoding:
    df1[i]=le.fit_transform(df1[i])
df1.head()
```

```python
df['Age'].value_counts()
```

# Scaling

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
col=['PassengerId','Pclass','Sex','Age','SibSp','Parch','Fare','Embarked','title']
df2=df1[col]
scaler=MinMaxScaler()
#df2=df1.iloc[:,:]
df2[:]=scaler.fit_transform(df2[:])
df2.head()
```

```python
df2.columns
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=df2[['Pclass','Sex','Age','Embarked','title']]
y=df1['Survived']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=10)
```

```python
print('shape of x_train:',x_train.shape)
print('shape of y_train:',y_train.shape)
print('shape of x_test:',x_test.shape)
print('shape of y_test:',y_test.shape)
```

# Gradient Boosting Classification model

```python
from sklearn import metrics
from sklearn.ensemble import GradientBoostingClassifier
gbcl = GradientBoostingClassifier(n_estimators = 50, learning_rate = 0.05)
gbcl = gbcl.fit(x_train, y_train)
test_pred = gbcl.predict(x_test)
gbcl_score=gbcl.score(x_test , y_test)
print("The Gradient Boost classifier model accuracy is {}".format(gbcl_score))
print("The confusion matrix of Gradient Boost classifier model is")
print(metrics.confusion_matrix(y_test,y_test))
```

```python
y_pred=gbcl.predict_proba(x_test)
y_pred
```

# Ada Boosting Classification model

```python
from sklearn.ensemble import AdaBoostClassifier
my_ada_clf = AdaBoostClassifier()
```

```python
my_ada_clf.fit(x_train,y_train)
my_ada_preds = my_ada_clf.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, my_ada_preds))
```

# Extreme Gradient Boosting model

```python
from xgboost import XGBClassifier
```

```python
my_xgb_clf = XGBClassifier(booster = 'gbtree')
my_xgb_clf.fit(x_train, y_train)
```

```python

my_xgb_preds = my_xgb_clf.predict(x_test)
```
